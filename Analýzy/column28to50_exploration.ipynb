## Data preparation

```python
data["bathrooms_text"][:50]
```

```python
import numpy as np
import re

#Getting the numeric values from bathrooms_text
def extract_numerical_value(text):
    # Regular expression pattern to match numerical values
    pattern = r'(\d+(\.\d+)?)'  # Matches one or more digits, possibly followed by a decimal point and more digits
    matches = re.findall(pattern, str(text))
    if matches:
        return float(matches[0][0])  # Extracting the numerical value and converting it to float
    else:
        return np.nan  # Return NaN if no numerical value is found

data["bathrooms_numeric"] = data['bathrooms_text'].apply(extract_numerical_value)
data["bathrooms_numeric"] = data["bathrooms_numeric"].astype(float)
data["bathrooms_numeric"][:50]

#We could also create categories such as "0 baths", "1 bath", "2 baths", "3+ baths"
```

```python
# Creating a binary column which indicates whether bathroom is shared (1) or not (0)
def extract_shared(text):
    if pd.isnull(text):
        return 0  # Return 0 if the value is NaN
    elif 'shared' in text.lower():
        return 1  # Return 1 if the word "shared" is found
    else:
        return 0  # Return 0 otherwise
data["bathrooms_shared"] = data['bathrooms_text'].apply(extract_shared)
data["bathrooms_shared"] = data['bathrooms_shared'].astype('category')
data["bathrooms_shared"].value_counts()
```

```python
#Creating a dataframe with only desired columns
columns_to_keep = ["neighbourhood_cleansed","room_type", "accommodates", "bathrooms_numeric", "bedrooms", "beds", "amenities", "price", "minimum_nights", "maximum_nights", "bathrooms_shared"]
data_relevant = data[columns_to_keep]
```

```python
#Removing $ symbol and comma from price to be able to change its datatype to float
data_relevant["price"] = data_relevant["price"].str.replace('$', '')
data_relevant["price"] = data_relevant["price"].str.replace(',', '')
```

```python
data_relevant["price"] = data_relevant["price"].astype(float)
data_relevant["price"][:10]
```

```python
#removing outliers
data_filtered = data_relevant[data_relevant['price'] < data_relevant['price'].quantile(0.95)]
```
